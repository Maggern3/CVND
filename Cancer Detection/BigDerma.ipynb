```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils, datasets
train_on_gpu = torch.cuda.is_available()
if train_on_gpu: 
    print('GPU/CUDA works! Success!')
else:
    print('CUDA doesnt work try updating drivers/OS')
    
torch.cuda.current_device()
torch.cuda.empty_cache()
```

```python
# import the data set with a DataLoader
data_transform = transforms.Compose([transforms.Resize(225),
                                     transforms.CenterCrop(224),
                                     transforms.ToTensor(),
                                     transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])])
test_data_transform = transforms.Compose([transforms.Resize(225),
                                          transforms.CenterCrop(224),
                                          transforms.ToTensor(),
                                          transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])])

data_dir = 'D:/dldata/dermatologist-ai/'
model_dir = 'saved_models/'
dataset = datasets.ImageFolder('D:/dldata/dermatologist-ai/train/train', transform=data_transform)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)
test_dataset = datasets.ImageFolder('D:/dldata/dermatologist-ai/valid', transform=test_data_transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
full_test_dataset = datasets.ImageFolder('D:/dldata/dermatologist-ai/test/test', transform=test_data_transform)
full_test_loader = DataLoader(full_test_dataset, batch_size=32, shuffle=False)
```

```python
import torch.nn as nn
import torch.nn.functional as F
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 64, 3, stride=2, padding=1)
        self.batchnorm1 = nn.BatchNorm2d(64)
        self.conv2 = nn.Conv2d(64, 64, 3, padding=1)
        self.batchnorm2 = nn.BatchNorm2d(64)
        self.conv2_1 = nn.Conv2d(64, 64, 3, padding=1)
        self.batchnorm2_1 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 128, 3, padding=1)
        self.batchnorm3 = nn.BatchNorm2d(128)
        self.conv3_1 = nn.Conv2d(128, 128, 3, padding=1)
        self.batchnorm3_1 = nn.BatchNorm2d(128)
        self.conv3_2 = nn.Conv2d(128, 128, 3, padding=1)
        self.batchnorm3_2 = nn.BatchNorm2d(128)
        self.conv4 = nn.Conv2d(128, 256, 3, padding=1)
        self.batchnorm4 = nn.BatchNorm2d(256)
        self.conv4_1 = nn.Conv2d(256, 256, 3, padding=1)
        self.batchnorm4_1 = nn.BatchNorm2d(256)
        self.conv4_2 = nn.Conv2d(256, 256, 3, padding=1)
        self.batchnorm4_2 = nn.BatchNorm2d(256)
        self.conv5 = nn.Conv2d(256, 512, 3, padding=1)
        self.batchnorm5 = nn.BatchNorm2d(512)
        self.conv5_1 = nn.Conv2d(512, 512, 3, padding=1)
        self.batchnorm5_1 = nn.BatchNorm2d(512)
        self.conv5_2 = nn.Conv2d(512, 512, 3, padding=1)
        self.batchnorm5_2 = nn.BatchNorm2d(512)
        
        self.fc1 = nn.Linear(4608, 1024)
        self.fc2 = nn.Linear(1024, 512)
        self.fc3 = nn.Linear(512, 3)
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(p=0.35)
        
    def forward(self, x):
        x = self.dropout(F.relu(self.batchnorm1(self.conv1(x))))
        x = self.dropout(F.relu(self.batchnorm2(self.conv2(x))))
        x = self.dropout(F.relu(self.batchnorm2_1(self.conv2_1(x))))
        x = self.dropout(self.pool(F.relu(self.batchnorm3(self.conv3(x)))))
        x = self.dropout(F.relu(self.batchnorm3_1(self.conv3_1(x))))
        x = self.dropout(F.relu(self.batchnorm3_2(self.conv3_2(x))))
        x = self.dropout(F.relu(self.batchnorm4(self.conv4(x))))
        x = self.dropout(F.relu(self.batchnorm4_1(self.conv4_1(x))))
        x = self.dropout(self.pool(F.relu(self.batchnorm4_2(self.conv4_2(x)))))
        x = self.dropout(self.pool(F.relu(self.batchnorm5(self.conv5(x)))))
        x = self.dropout(self.pool(F.relu(self.batchnorm5_1(self.conv5_1(x)))))
        x = self.dropout(self.pool(F.relu(self.batchnorm5_2(self.conv5_2(x)))))
        
        x = x.view(x.shape[0], -1)
        x = self.dropout(self.fc1(x))
        x = self.dropout(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
import torch.optim as optim
import torch.optim.lr_scheduler as lr_scheduler

criterion = nn.CrossEntropyLoss()

net = Net()
print(net)
net.cuda()

optimizer = optim.Adam(net.parameters(), lr=0.001)

lr_scheduler = lr_scheduler.ExponentialLR(optimizer, 0.98)

training_losses, validation_losses = [], []
smallest_validation_loss = 1.0
current_lr = 0.0
n_epochs = 60
for epoch in range(n_epochs):    
    running_loss = 0.0
    #lr_scheduler.step()
    current_lr = lr_scheduler.get_lr()
    for batch_i, data in enumerate(train_loader):
        net.train()
        images, labels = data
        images = images.cuda()
        labels = labels.cuda()
        optimizer.zero_grad()
        output = net(images)
        loss = criterion(output, labels)        
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    with torch.no_grad():
        net.eval()
        running_loss = running_loss / len(train_loader)
        training_losses.append(running_loss) 
        validation_loss = 0.0
        for batch_i, data in enumerate(test_loader):
            images, labels = data
            images = images.cuda()
            labels = labels.cuda()
            output = net(images)
            validation_loss += criterion(output, labels)
        validation_loss = validation_loss / len(test_loader)
        validation_losses.append(validation_loss)
        if validation_loss < smallest_validation_loss:
            smallest_validation_loss = validation_loss
            print('Saving model at epoch ', epoch+1)            
            model_name = 'weights_modelv8_2_e'+str(epoch+1)+'.pt'
            torch.save(net.state_dict(), model_dir+model_name)
        print('Epoch: {} Training Loss: {:6.4f} Validation Loss: {:6.4f} lr: {}'.format(epoch+1, running_loss, validation_loss, current_lr))
        
```
